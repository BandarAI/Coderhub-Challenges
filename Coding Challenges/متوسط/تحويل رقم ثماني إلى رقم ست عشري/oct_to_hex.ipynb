# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل متغير من نوع int يعبر عن قيمة ثمانية octal number، ثم تقوم الدالة بإرجاع النتيجة بعد التحويل الى قيمة ست عشرية hexadecimal number بنوع string
</div>


# Solution

```python
from typing import List
def oct_to_hex(octal_number: int) -> str:
    return hex(int(str(octal_number), 8)).upper()[2:]
```

```python
oct_to_hex(octal_number = 5252)
```

```python
oct_to_hex(octal_number = 777)
```

```python
oct_to_hex(octal_number = 123)
```

```python
oct_to_hex(octal_number = 1000)
```
