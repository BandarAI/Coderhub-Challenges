# Problem Description

<div dir="rtl" style="text-align: center;">
إذا كانت الارتفاعات [1, 2, 3, 2, 3, 4, 1]:

من 1→2→3 (متزايد) طول=3

ثم انكسرت عند 3→2

بدأت سلسلة جديدة 2→3→4 (متزايد) طول=3
أطول سلسلة وجدتها هي 3 أيام متواصلة من الزيادة.
</div>


# Solution

```python
from typing import List
def longestClimbSegment(heights: List[int]) -> int:
    counter = 1
    longest_counter = 1
    for height in range(1, len(heights)):
        if heights[height] > heights[height-1]:
            counter += 1
            longest_counter = max(longest_counter, counter)
        else:
            counter = 1
    return longest_counter
            

```

```python
longestClimbSegment(heights = [1, 2, 3, 2, 3, 4, 1])
```

# Computational Complexity

## Time Complexity O(n)
## Space Complexity O(1)
