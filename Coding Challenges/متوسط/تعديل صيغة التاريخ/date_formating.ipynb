# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل قيمة نصية string تُمثل تاريخ اليوم، تقوم الدالة بإرجاع تاريخ اليوم بثلاث تنسيقات مختلفة.
</div>


## Solution

```python
from typing import List
def date_formating(date: str) -> str:
    y, m, d = date.split("-")
    return f"{d}-{m}-{y}"
```

```python
date_formating(date = '2020-01-01')
```

```python
date_formating(date = '2021-01-01')
```

```python
date_formating(date = '1954-09-01')
```

```python
date_formating(date = '2000-12-07')
```

### Computational Complexity

#### Time Complexity $O(1)$
#### Space Complexity $O(1)$
