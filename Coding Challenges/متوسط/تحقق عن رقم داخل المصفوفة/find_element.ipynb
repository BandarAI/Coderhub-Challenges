# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل مصفوفة array من نوع integer ، وتقوم بالتحقق من integer محدد مما إذا كان ضمن عناصر الـمفصوفه array أم لا، ثم تقوم بإرجاع النتيجة من نوع Boolean .
</div>


# Solution

```python
from typing import List
def find_element(elements_array: List[int], element: int) -> bool:
    return element in elements_array
```

```python
find_element(elements_array = [2, 12, 34, 2, 3, 4, 5], element = 4)
```

# Computational Complexity

## Time Complexity O(n)
## Space Complexity O(1)
