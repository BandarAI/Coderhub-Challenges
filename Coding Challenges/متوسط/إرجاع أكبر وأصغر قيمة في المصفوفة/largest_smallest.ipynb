# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل مصفوفة array من نوع integer وتقوم بإرجاع أكبر عدد وأصغر عدد في المصفوفة.
</div>


# Solution

```python
from typing import List
def largest_smallest(array_values: List[int]) -> List[int]:
    return [max(array_values), min(array_values)]
```

```python
largest_smallest(array_values = [5, 1, 2, 4, 9, 10, 200])
```

# Computational complexity

## Time Complexity O(n)
## Space Complexity O(1)
