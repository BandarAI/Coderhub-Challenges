# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة تستقبل مصفوفة من القيم النصية، تقوم تلك الدالة بمقارنة القيم النصية وإرجاع true في حال كان عدد characters لجميع النصوص متطابق، وإرجاع false في حال خلاف ذلك. اذا كانت المصفوفة فارغة أو تحتوي على عنصر واحد فقط، قم بإرجاع القيمة false.
</div>


# Solution

```python
from typing import List
def str_len_comparison(words: List[str]) -> bool:
    return len(words) > 1 and all(len(word) == len(words[0]) for word in words)
```

```python
str_len_comparison(words = [""])
```

```python
str_len_comparison(words = ['A'])
```

```python
str_len_comparison(words = ['Khalid', 'Sultan', 'AlDana', 'Johrah', 'Ghadah'])
```

# Computational complexity

## Time Complexity O(n)
## Space Complexity O(1)
