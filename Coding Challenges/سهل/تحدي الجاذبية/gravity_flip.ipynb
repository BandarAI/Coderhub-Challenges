# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function: تقوم الدالة باستقبال مصفوفة أعداد صحيحة من نوع integer. كل خلية في هذه المصفوفة تمثل عدد المكعبات في هذا العامود، كما تقوم الدالة بالتحقق من عدد المكعبات في كل عامود وإحالتها إلى الأعمدة الأصغر. ‎</div>


# Solution

```python
from typing import List
def gravity_flip(columns: List[int]) -> List[int]:
    return sorted(columns)
```

```python
gravity_flip(columns = [4, 5, 6, 3, 2])
```

# Computational complexity

## Time Complexity O(n lon n)
## Space Complexity O(n)
