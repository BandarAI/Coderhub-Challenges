# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل عدد صحيح integer وتحدد فيما إذا كانت القيمة المرسلة أصغر من أو أكبر من أو تساوي الصفر. تقوم الدالة بإرجاع قيمة نصية string تُعبّر عن حالة القيمة المرسلة.
</div>


# Solution

```python
from typing import List
def less_or_more_than_zero(number: int) -> str:
    return "Greater than zero" if number > 0 else "Equal to zero" if number == 0 else "Less than zero"

```

```python
less_or_more_than_zero(number = 10)
```

```python
less_or_more_than_zero(number = 0)
```

```python
less_or_more_than_zero(number = -10)
```

# Computational Complexity

## Time Complexity: $O(n)$
## Space Complexity: $O(1)$
