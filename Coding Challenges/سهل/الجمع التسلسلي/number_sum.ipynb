## 1- Intern Tom Solution

```python
from typing import List
def number_sum(num: int) -> int:
    return sum(range(1, num+1))
```

```python
number_sum(num = 3)
```

### Computational Complexity

#### Time Complexity O(n)
#### Space Complexity O(1)

## 2- Engineer Tom Solution

```python
from typing import List
def number_sum(num: int) -> int:
    return num * (num + 1) // 2
```

```python
number_sum(num = 3)
```

### Computational Complexity

#### Time & Space Complexity O(1)
