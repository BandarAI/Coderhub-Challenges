# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل عدد من نوع integer ، وتقوم الـدالة function بطباعة الأرقام من صفر إلى نفس العدد المدخل كقيمة نصية من نوع string</div>


# Solution

```python
from typing import List
def numbers_range(number: int) -> str:
    return " ".join(str(n) for n in range(number+1))
```

```python
numbers_range(number = 4)
```

# Computational Complexity

## Time & Space Complexity O(n)
