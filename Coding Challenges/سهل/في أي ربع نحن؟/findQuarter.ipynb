# Problem Description

<div dir="rtl" style="text-align: right;">
تخيّل أنك تعمل في شركة تعتمد خططها على فصول السنة المالية، والتي تنقسم إلى أربعة أرباع:

    الربع الأول (Q1): من يناير إلى مارس (الشهور 1-3).

    الربع الثاني (Q2): من أبريل إلى يونيو (الشهور 4-6).

    الربع الثالث (Q3): من يوليو إلى سبتمبر (الشهور 7-9).

    الربع الرابع (Q4): من أكتوبر إلى ديسمبر (الشهور 10-12).

يأتي صديقك ويقول لك:

    "لدي شهر نوفمبر، في أي ربع يقع؟"

تجيب بسرعة:

    "نوفمبر يقع في الربع الرابع!"

ثم يسألك عن شهر فبراير، فترد عليه:

    "فبراير يقع في الربع الأول!"

وهكذا، مهمتك هي بناء دالة تأخذ رقم الشهر وتخبرك بأي ربع ينتمي هذا الشهر.

مهمتك الآن هي تصميم دالة تقوم بتحديد الربع الذي ينتمي إليه أي شهر يتم إدخاله.
</div>


## Solution

```python
from typing import List
def findQuarter(month: int) -> int:
    if 1 <= month <= 3:
        return 1
    if 4 <= month <= 6:
        return 2
    if 7 <= month <= 9:
        return 3
    if 10 <= month <= 12:
        return 4
```

```python
findQuarter(month = 1)
```

```python
findQuarter(month = 2)
```

```python
findQuarter(month = 3)
```

```python
findQuarter(month = 4)
```

#### Time Complexity $O(1)$
#### Space Complexity $O(1)$
