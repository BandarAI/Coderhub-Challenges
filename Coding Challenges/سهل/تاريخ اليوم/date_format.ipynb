# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل قيمة نصية string تُمثل تاريخ اليوم، تقوم الدالة بإرجاع تاريخ اليوم بثلاث تنسيقات مختلفة.
</div>


## Solution

```python
from typing import List
def date_format(date: str) -> str:
    y, m, d = date.split("/")

    return f"{y}/{m}/{d} | {y}-{m}-{d} | {m}/{d}/{y}"
```

```python
date_format(date = '2020/1/1')
```

```python
date_format(date = '2019/12/28')
```

```python
date_format(date = '2010/10/30')
```

```python
date_format(date = '2013/11/29')
```

#### Time Complexity $O(n)$
#### Space Complexity $O(n)$
