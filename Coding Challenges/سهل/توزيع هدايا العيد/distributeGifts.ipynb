# Problem Description

<div dir="rtl" style="text-align: right;">

إذا كانت العائلات كما يلي:

    العائلة الأولى: 4 أفراد، مجموع الهدايا 20

    العائلة الثانية: 5 أفراد، مجموع الهدايا 25
    فإن النتيجة ستكون:

    العائلة الأولى: 5 هدايا لكل فرد

    العائلة الثانية: 5 هدايا لكل فرد


</div>


# Solution

```python
from typing import List
def distributeGifts(familySizes: List[int], totalGifts: List[int]) -> List[int]:
    return [int(gifts / sizes) for gifts, sizes in zip(totalGifts, familySizes)]
```

```python
distributeGifts(familySizes = [4], totalGifts = [20])
```

```python
distributeGifts(familySizes = [5], totalGifts = [25])
```

```python
distributeGifts(familySizes = [4, 5], totalGifts = [20, 25])
```

```python
distributeGifts(familySizes = [2, 3], totalGifts = [8, 9])
```
