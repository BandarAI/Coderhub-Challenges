# Problem Description

<div dir="rtl" style="text-align: right;">
يُطلب منك كتابة دالة تُدعى calculateSaudiFlagScore، حيث تستقبل الدالة مصفوفة من العناصر الرمزية (على سبيل المثال: "Green" أو "Sword" أو "Arabic Calligraphy" أو "Palm") وتُعيد مجموع النقاط بناءً على قيمة كل عنصر.

كيفية عمل الدالة:

    يُعطى لكل عنصر من العناصر رموز محددة مسبقاً:

        "Green": 10 نقطة.

        "Sword": 15 نقطة.

        "Arabic Calligraphy": 20 نقطة.

        "Palm": 5 نقطة.

        أي عنصر آخر: 0 نقطة.

    تُحسب الدالة إجمالي النقاط بجمع نقاط جميع العناصر.

    تُعاد النتيجة كعدد صحيح يمثل مجموع النقاط.

مثال :

إذا كانت المصفوفة المدخلة هي:

["Green", "Sword", "Arabic Calligraphy"]

فإن:

    النقاط = 10 (Green) + 15 (Sword) + 20 (Arabic Calligraphy) = 45.

    النتيجة المتوقعة: 45
</div>


# Solution

```python
from typing import List
def calculateSaudiFlagScore(elements: List[str]) -> int:
    flag_dict = {
        "Green":10,
        "Sword":15,
        "Arabic Calligraphy":20,
        "Palm":5
    }

    return sum(flag_dict[element] for element in elements if element in flag_dict)
```

```python
calculateSaudiFlagScore(elements = ['Green'])
```

```python
calculateSaudiFlagScore(elements = ['Sword'])
```

```python
calculateSaudiFlagScore(elements = ['Arabic Calligraphy'])
```

```python
calculateSaudiFlagScore(elements = ['Green', 'Sword'])
```
