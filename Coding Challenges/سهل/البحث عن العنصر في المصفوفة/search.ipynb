# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل قيمة نصية وحرف. تقوم الدالة بالبحث عن أول index مطابق للحرف المُدخل وإرجاع رقم index الخاص بذلك الحرف. تقوم الدالة بإرجاع القيمة -1 في حال عدم وجود تطابق.</div>


# Solution

## 1- .find()

```python
from typing import List
def search(word: str, character: str) -> int:
    return word.find(character)
```

```python
search(word = 'cloud', character = 'u')
```

```python
search(word = 'sofa', character = 'k')
```

## 2- .index()

```python
from typing import List
def search(word: str, character: str) -> int:
    try:
        return word.index(character)
    except ValueError:
        return -1
```

```python
search(word = 'cloud', character = 'u')
```

```python
search(word = 'sofa', character = 'k')
```

# Computational Complexity

## Time Complexity O(n)
## Space Complexity O(1)
