# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل قيمتين نصية من نوع string وتقوم بالتحقق فيما إذا كانتا متشابهتين أم غير متشابهتين.
</div>


# Solution

```python
from typing import List
def is_same(name1: str, name2: str) -> str:
    return "متشابهتين" if name1 == name2 else "غير متشابهتين"
```

```python
is_same(name1 = 'Anas', name2 = 'anas')
```

# Computational Complexity

## Space Complexity: O(n)
## Time Complexity: O(1)
