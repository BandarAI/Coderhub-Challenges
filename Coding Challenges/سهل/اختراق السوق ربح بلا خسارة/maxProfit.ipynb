# Problem Description

<div dir="rtl" style="text-align: right;">
في عالم الأمن السيبراني، تتعامل الشركات مع معلومات حساسة لا تقدر بثمن، تمامًا كما يتعامل المستثمرون مع أسعار الأسهم في الأسواق المالية. في أحد مؤتمرات بلاك هات، ظهرت أداة حديثة أثارت فضول الخبراء؛ إذ تعتمد على أربعة عوامل أساسية لتحليل حركة السوق وكشف الأنماط المخفية التي قد تقود إلى أرباح كبيرة بشكل غير متوقع.

ضمن هذا السياق، تم تكليفك بمهمة أمنية دقيقة: بناء خوارزمية قادرة على تحديد أفضل توقيت لشراء وبيع الأسهم لتحقيق أكبر ربح ممكن. التحدّي الحقيقي ليس في فهم البيانات فقط، بل في سرعة اتخاذ القرار—فكل ثانية قد تعني فرصة تُغتنم أو ضياع مكسب محتمل.

هل ستتمكن من قراءة المؤشرات الصحيحة وتحقيق أقصى ربح دون المساس بسرية المعلومات؟ أم ستفلت الفرصة من بين يديك

    المهمة:

        لديك مصفوفة تحتوي على أسعار الأسهم اليومية.

        يجب أن تحدد اليوم المثالي للشراء واليوم المثالي للبيع لتحقيق أقصى ربح.

    القواعد:

        لا يمكنك البيع قبل الشراء.

        إذا لم يكن هناك ربح ممكن، فإن النتيجة هي 0.
</div>


## Solution

```python
from typing import List
def maxProfit(prices: List[int]) -> int:
    min_price = float('inf')
    max_profit = 0

    if not prices:
        return 0
    
    for price in prices:
        if price < min_price:
            min_price = price
        else:
            max_profit = max(max_profit, price - min_price)
    
    return max_profit
```

```python
maxProfit(prices = [7, 1, 5, 3, 6, 4])
```

```python
maxProfit(prices = [7, 6, 4, 3, 1])
```

```python
maxProfit(prices = [2, 4, 1])
```

```python
maxProfit(prices = [])
```

### Computational Complexity

#### Time Complexity $O(n)$
#### Space Complexity $O(1)$
