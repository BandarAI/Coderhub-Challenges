# Problem Description

<div dir="rtl" style="text-align: right;">
تخيل أنك مهندس معماري، وتحتاج إلى حساب مساحة متوازي الأضلاع لاستخدامه في تصميم بناء.
يجب أن يكون العرض (breadth) والارتفاع (height) قيمًا موجبة لضمان أن الأبعاد صحيحة.
إذا تم إدخال أي قيمة سالبة أو صفرية، ستتلقى رسالة خطأ تخبرك بأن "Breadth and height must be positive".

إذا كانت القيم المدخلة موجبة، يتم حساب مساحة متوازي الأضلاع باستخدام الصيغة:

    Area = Breadth × Height

قواعد التحدي:

    إذا كانت أي من القيمتين breadth أو height أقل من أو تساوي صفرًا، يجب طباعة الخطأ:
    "java.lang.Exception: Breadth and height must be positive".

    إذا كانت القيم موجبة، يتم حساب مساحة متوازي الأضلاع باستخدام الصيغة المذكورة أعلاه.
</div>


# Solution

```python
from typing import List
def calculateParallelogramArea(breadth: int, height: int) -> str:
    return "java.lang.Exception: Breadth and height must be positive" if breadth <= 0 or height <= 0 else str(breadth * height)
```

```python
calculateParallelogramArea(breadth = 1, height = 3)
```

```python
calculateParallelogramArea(breadth = -1, height = 2)
```

```python
calculateParallelogramArea(breadth = 0, height = 7)
```
