# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل مصفوفة array من نوع integer. تقوم الدالة بإرجاع العناصر المتشابهة في المصفوفة.
</div>


# Solution

```python
from typing import List
def get_duplicate_elements(arr: List[int]) -> List[int]:
    seen = set()
    duplicates = set()
    result = []

    for element in arr:
        if element in seen:
            if element not in duplicates:
                result.append(element)
                duplicates.add(element)
        else:
            seen.add(element)

    return result
```

```python
get_duplicate_elements(arr = [2, 3, 2, 3])
```

```python
get_duplicate_elements(arr = [10, 5, 9, 5])
```

```python
get_duplicate_elements(arr = [6, 3, 12, 12])
```

```python
get_duplicate_elements(arr = [3, -9, -9, -4])
```

## Time and Space Complexity O(n)
