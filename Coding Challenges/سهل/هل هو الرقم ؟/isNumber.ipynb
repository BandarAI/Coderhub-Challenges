# Problem Description

<div dir="rtl" style="text-align: right;">
في عالم السيبرانيين، حيث تُستخدم الأرقام كرموز سرية في كل مكان، تصلك رسالة تحتوي على سلسلة نصية غامضة. مهمتك هي التحقق مما إذا كانت هذه السلسلة تمثل رقمًا صالحًا أو مجرد تشويش لا معنى له.

لديك سلسلة نصية تحتوي على رموز وأرقام، ومهمتك هي تحديد ما إذا كانت هذه السلسلة تمثل رقمًا صالحًا. يمكن أن تكون الأرقام بأشكال مختلفة: أرقام صحيحة، أرقام عشرية، أو أرقام علمية تحتوي على أسس.
ما الذي يجعل السلسلة رقمًا صالحًا؟

    يمكن أن تحتوي على أرقام فقط.

    قد تبدأ بـ + أو - للإشارة إلى علامة الرقم.

    يمكن أن تحتوي على فاصلة عشرية (.) أو حتى جزء أسي (e أو E).

    ما الذي يجعل السلسلة غير صالحة؟

        وجود أحرف غير رقمية (مثل a أو z).

        تنسيقات غير صحيحة مثل 1e, e3, أو 1.2.3.

</div>


# Solution

```python
from typing import List
def isNumber(s: str) -> bool:
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
isNumber(s = '0')
```

```python
isNumber(s = '-123.456e789')
```

```python
isNumber(s = '53.5e93')
```

```python
isNumber(s = 'abc')
```
