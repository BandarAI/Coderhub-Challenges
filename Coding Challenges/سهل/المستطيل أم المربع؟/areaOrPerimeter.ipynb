# Problem Description

<div dir="rtl" style="text-align: right;">
تخيّل أنك بنّاء تعمل على تصميم أشكال هندسية، وكل شكل يأتيك بمعلومتين: الطول و العرض.
مهمتك هي تحديد الشكل ثم حساب ما يحتاجه:

    إذا كان الطول والعرض متساويين، فهذا يعني أن الشكل مربع، وستقوم بحساب مساحته.

    إذا كان الطول والعرض غير متساويين، فهذا يعني أن الشكل مستطيل، وستقوم بحساب محيطه.

مثلاً، صديقك البنّاء يعطيك القياسات التالية:

    "لديّ قطعة خشب طولها 3 أمتار وعرضها 3 أمتار، ما الذي يجب أن أفعله؟"

    ترد عليه: "إنها مربع! مساحتها هي 3×3=93 × 3 = 93×3=9 أمتار مربعة."

ثم يعطيك قياسات جديدة:

    "وهنا قطعة أخرى طولها 6 أمتار وعرضها 10 أمتار."

    ترد عليه: "إنها مستطيل! محيطها هو 2×(6+10)=322 × (6 + 10) = 322×(6+10)=32 مترًا."

مثال:

    إذا كانت الأبعاد 3 و 3:

        الشكل مربع → المساحة = 3×3=93 × 3 = 93×3=9.

    إذا كانت الأبعاد 6 و 10:

        الشكل مستطيل → المحيط = 2×(6+10)=322 × (6 + 10) = 322×(6+10)=32.
</div>


# Solution

```python
from typing import List
def areaOrPerimeter(length: int, width: int) -> int:
    return length * width if length == width else (length + width)*2
```

```python
areaOrPerimeter(length = 3, width = 3)
```

```python
areaOrPerimeter(length = 6, width = 10)
```
