# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل مصفوفة من نوع integer، تقوم الدالة بجمع الأعداد الزوجية فقط وإرجاع النتيجة كعدد صحيح integer.
</div>


# Solution

## Method 1 - Modulo Check

```python
from typing import List
def sum_even(arr: List[int]) -> int:
    return sum(i for i in arr if i % 2 == 0)
```

```python
sum_even(arr = [11, 0, 5, 22])
```

## Method 2 - Bitwise Check

```python
from typing import List
def sum_even(arr: List[int]) -> int:
    return sum(i for i in arr if (i & 1) == 0)
```

```python
sum_even(arr = [11, 0, 5, 22])
```

# Computational Complexity

## Time Complexity O(n)
## Space Complexity O(1)
