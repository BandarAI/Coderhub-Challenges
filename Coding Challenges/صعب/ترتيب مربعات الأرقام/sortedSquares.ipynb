# Problem Description

<div dir="rtl" style="text-align: right;">
تخيل أنك تعمل على مشروع لتحليل بيانات درجات الحرارة اليومية، وتمثل الأرقام السالبة درجات البرودة والأرقام الموجبة درجات الحرارة المرتفعة. هدفك هو تحويل هذه القيم إلى درجات طاقة باستخدام التربيع، ثم ترتيب النتائج لمعرفة توزيع الطاقة بشكل منظم.

المدخل :
nums = [-4, -1, 0, 3, 10]

الخطوات :
1 قم بتربيع كل عنصر:
مربعات الأرقام = [16, 1, 0, 9, 100]

2 قم بترتيب المربعات تصاعديًا:

[0, 1, 9, 16, 100]

الاخراج :
[0, 1, 9, 16, 100]


لكن التحدي هنا هو أن القيم السالبة تصبح موجبة عند التربيع، مما يجعل ترتيب النتائج يتطلب نهجًا ذكيًا.
</div>


## Solution

```python
from typing import List
def sortedSquares(nums: List[int]) -> List[int]:
    return sorted([(n) ** 2 for n in nums])
```

```python
sortedSquares(nums = [-4, -1, 0, 3, 10])
```

```python
sortedSquares(nums = [-7, -3, 2, 3, 11])
```

```python
sortedSquares(nums =  [-5, -3, -1, 0, 1, 4, 6])
```

```python
sortedSquares(nums = [0, 1, 2, 3, 4])
```

### Computational Complexity

#### Time Complexity: $O(nlogn)$
#### Space Complexity: $O(n)$
