# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل قيمة نصية من نوع string ، تقوم الدالة بفصل الكلمات عند ملاحظة حرف كبير بوضع مسافه وتحويله لحرف صغير، ثم قم بارجاع النتيجة من نوع string
</div>


## Solution

```python
from typing import List
def cap_space(txt: str) -> str:
    word = []
    current = ""
    
    for ch in txt:
        if ch.isupper() and current:
            word.append(current)
            current = ch
        else:
            current += ch
    
    word.append(current)
    return " ".join(word).lower()
```

```python
cap_space(txt = 'wePlayTennis')
```

```python
cap_space(txt = 'iLikeCats')
```

```python
cap_space(txt = 'computerScience')
```

```python
cap_space(txt = 'thankYou')
```

### Computational Complexity

#### Time Complexity: $O(n)$
#### Space Complexity: $O(n)$
