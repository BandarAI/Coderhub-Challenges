# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل قيمة نصية string تعبر عن الساعة كنظام 12 أو 24 ساعة و تقوم بالتحويل بينها ، و قم بإلحاق نظام 12 ساعة بـ am أو pm ثم قم بارجاع النتيجة من نوع string
</div>


## Solution

```python
from typing import List
def convert_time(time: str) -> str:
    time = time.strip().lower()

    if 'am' in time or 'pm' in time:
        t, period = time.split()
        hour, minute = map(int, t.split(':'))

        if period == 'am':
            if hour == 12:
                hour = 0
        else:  # pm
            if hour != 12:
                hour += 12

        return f"{hour}:{minute:02d}"

    else:
        hour, minute = map(int, time.split(':'))

        if hour == 0:
            return f"12:{minute:02d} am"
        elif hour < 12:
            return f"{hour}:{minute:02d} am"
        elif hour == 12:
            return f"{hour}:{minute:02d} pm"
        else:
            return f"{hour - 12}:{minute:02d} pm"
```

```python
convert_time(time = '10:30 am')
```

```python
convert_time(time = '7:13 pm')
```

```python
convert_time(time = '19:00')
```

```python
convert_time(time = '12:00 am')
```

### Computational Complexity

#### Time Complexity: $O(1)$
#### Space Complexity: $O(1)$
