# Problem Description

<div dir="rtl" style="text-align: right;">
اكتشاف النطاق الأول والأخير لهذا المفتاح داخل السلسلة

مثال :
الادخال :
"name": "nums",

"value": [2, 3, 3, 3, 4, 5, 6]

الاخراج:

[1, 3]

التفسير:
في القائمة [2, 3, 3, 3, 4, 5, 6]، الرقم 3 يظهر أول مرة عند الفهرس 1 وآخر مرة عند الفهرس 3. لذا يكون النطاق [1, 3].
 </div>


# Solution

```python
from typing import List
def searchRange(nums: List[int], target: int) -> List[int]:
    indexes = [index for index, value in enumerate(nums) if value == target]
    return [indexes[0], indexes[-1]] if indexes else [-1, -1]
```

```python
searchRange(nums = [5, 7, 7, 8, 8, 10], target = 8)
```

```python
searchRange(nums = [5, 7, 7, 8, 8, 10], target = 6)
```

```python
searchRange(nums = [], target = 0)
```

```python
searchRange(nums = [1], target = 1)
```
