# Solution

```python
from typing import List
def get_foundation_day_celebration(city: str) -> str:
    city_dict = {
        "الرياض": "الرياض: العرضة النجدية - المشلح والغترة - القرصان",
        "جدة": "جدة: الزفة الحجازية - الثوب الحجازي والشماغ - المعصوب",
        "الحدود الشمالية": "الحدود الشمالية: الدحة - البشت والعقال والثوب المرودن - المليحية",
        "حائل": "حائل: سامري حائل - البشت والعقال - كبيبة حائل",
        "الدمام": "الدمام: الفجري - الثوب والدقلة - المرقوق",
        "أبها": "أبها: الخطوة الجنوبية - الثوب الجنوبي - الحنيذ",

    }

    return city_dict.get(city, "معلومات غير متوفرة لهذه المدينة")
```

```python
get_foundation_day_celebration(city = 'جدة')
```

```python
get_foundation_day_celebration(city = 'زحل')
```
