# Problem Description

<div dir="rtl" style="text-align: right;">
    المدخلات:

        familySizes (int[]) – عدد أفراد كل عائلة.

        totalWeight (float) – الوزن الإجمالي للحم الأضحية بالكيلوغرام.

    ما يجب فعله:

        احسب إجمالي عدد الأشخاص: مجموع عناصر familySizes.

        احسب وزن اللحم لكل فرد: perPerson = totalWeight / totalPeople.

        لكل عائلة، احسب نصيبها: share = perPerson × familySize.

    الإخراج:
    مصفوفة من الأعداد (float[]) تمثل وزن اللحم الذي يذهب لكل عائلة.

مثال توضيحي:

    familySizes = [2, 3]

    totalWeight = 25.0

    المجموع الكلي للأشخاص = 2 + 3 = 5

    نصيب الفرد = 25.0 ÷ 5 = 5.0 كجم

    حصة العائلة الأولى = 5.0 × 2 = 10.0 كجم
    حصة العائلة الثانية = 5.0 × 3 = 15.0 كجم

الناتج: [10.0, 15.0]

</div>


## Solution

```python
from typing import List
def distributeSacrificeShares(familySizes: List[int], totalWeight: float) -> List[float]:
    total_people = sum(familySizes)
    per_person = totalWeight / total_people

    return [per_person * size for size in familySizes]
```

```python
distributeSacrificeShares(familySizes = [1], totalWeight = 10)
```

```python
distributeSacrificeShares(familySizes = [2], totalWeight = 10)
```

```python
distributeSacrificeShares(familySizes = [2, 3], totalWeight = 25)
```

```python
distributeSacrificeShares(familySizes = [1, 1, 1], totalWeight = 3)
```

### Computational Complexity

#### Time Complexity: $O(n)$
#### Space Complexity: $O(n)$
