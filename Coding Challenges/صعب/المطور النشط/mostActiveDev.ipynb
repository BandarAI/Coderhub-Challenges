# Problem Description

<div dir="rtl" style="text-align: right;">
في نهاية يوم مزدحم على المستودع، تريد معرفة من هو المطوّر الأكثر نشاطًا اعتمادًا على عدد الـ commits المسجّلة بأسمائهم. لديك قائمة بأسماء المؤلفين بحسب تسلسل الـ commits (قد تظهر الأسماء عدة مرات). المطلوب إرجاع اسم المطوّر صاحب أكبر عدد من الـ commits.
إذا حصل تعادل في العدد، أرجِع الاسم الأصغر أبجديًا (lexicographically). إذا كانت القائمة فارغة، أرجِع سلسلة فارغة "".

مثال توضيحي:
authors = ["Ali", "Sara", "Ali", "Omar"]

    Ali: 2

    Sara: 1

    Omar: 1
    النتيجة ⇒ "Ali".
</div>


## Solution

```python
from typing import List
def mostActiveDev(authors: List[str]) -> str:
    devs = []
    count = {}

    if not authors:
        return ""

    for author in authors:
        count[author] = count.get(author, 0) + 1
    
    max_commits = max(count.values())
    for author, freq in count.items():
        if freq == max_commits:
            devs.append(author)
    return min(devs)
```

```python
mostActiveDev(authors = ['Ali', 'Sara', 'Ali'])
```

```python
mostActiveDev(authors = ['Dev1', 'Dev2', 'Dev2', 'Dev1'])
```

```python
mostActiveDev(authors = ['x'])
```

```python
mostActiveDev(authors = [])
```

### Computational Complexity

#### Time Complexity: $O(n)$
#### Space Complexity: $O(n)$
